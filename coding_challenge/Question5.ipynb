```python
def calculate_si(P, R, T):
    SI = (P * R * T) / 100
    return SI

P = float(input("Enter principal"))
R = float(input("Enter rate of interest"))
T = float(input("Enter time"))

result = calculate_si(P, R, T)
print("Simple Interest is", result)
```
