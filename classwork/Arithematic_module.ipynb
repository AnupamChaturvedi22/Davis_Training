```python
def sum(num1,num2):
   return num1 + num2

def subtract(num1,num2):
   return num1 - num2

def product(num1,num2):
   return num1 * num2

def division(num1,num2):
   return num1 / num2

def modulus(num1,num2):
   return num1 % num2
```
