```python
# count vowels from existing file
    f = open("firstfile.txt", "r")

    if f.mode == "r":
        data = f.read()

        print("File Content:")
        print(data)
        print("---------------------------")

        # Count vowels
        count = 0
        for ch in data:
            if ch.lower() in "aeiou":
                count += 1

        print("Number of vowels:", count)

        f.close()

    else:
        print("Unable to open the file")
```
