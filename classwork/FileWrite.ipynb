```python
# program for writting data into the file
# opening file for write operation
filev = open("firstfile.txt","w")
# blank list for sentences
sentences = []
# writing five sentences into the file
print("Enter any five sentence :")
for x in range(5):
  # input data from user
  sentence = input()
  # inserting the sentence into the list
  sentences.append(sentence)
  print("-----------------------------")

# writing data into the list
filev.writelines(sentences)
print("data succesfully written")
# closing the file
filev.close()
```
