```python
# Take a number from the user
n = int(input("Enter number: "))

# Loop from 1 to the entered number
for i in range(1, n+1):

    # Check if i is a factor of n and also even
    if n % i == 0 and i % 2 == 0:

        # Print the even factor
        print(i)
```
